```python
from os import remove
from faker import Faker
listafalsa = Faker()
print("Inserir quantos nomes serão listados:")
x=int(input())
remove("Teste.txt")

with open('Teste.txt', 'x'):
  for i in range(x):
    print(listafalsa.name(), listafalsa.numerify())
    with open('Teste.txt', 'a') as escrever:
      escrever.write(listafalsa.name())
      escrever.write("  ")
      escrever.write(listafalsa.numerify())
      escrever.write("  ")
```

```python
#mapear o arquivo txt de forma a identificar nomes e numeros
pessoas = []

with open('Teste.txt', 'r') as read:
  Dados = read.readlines()

  for linha in Dados:
    for i in linha.split('  '):
      pessoas.append(i)


print(pessoas)
#separando as listas entre nomes e notas
a=[]
b=[]
for i in range(len(pessoas)):
  if i%2==0:
   a.append(pessoas[i])
  else:
   b.append(pessoas[i])

print(a)
print(b)
c = [int(val) for val in b]
print(c)

```

```python

from matplotlib import pyplot as plt
import numpy as np


plt.hist(c, bins=50)
plt.title('Histograma das pontuações')
plt.ylabel('Alunos por nota')
plt.xlabel('Notas')
plt.show()
```

```python
#criação de grafico em worldcloud, a partir do arquivo teste.txt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

text = open("Teste.txt", 'r').read()
stopwords = STOPWORDS

wc = WordCloud(
    background_color='Black',
    stopwords=STOPWORDS,
    height = 800, width = 800)
wc.generate(text)

wc.to_file('teste.png')
```
